```python
#CAPA COMUNAS

highlight_function = lambda x: {'fillColor': '#000000', 
                                'color':'#000000', 
                                'fillOpacity': 0.50, 
                                'weight': 0.1}

#escala
escala_de_color = secciones_map.copy() #hagamos una copia de circuitos
vals = escala_de_color['cantidad_arboles'].describe()

#ordeno los numeros (vals) para generar la escala de colores
if vals['min'] == 0:
    ordenada = [vals['25%']-vals['std'],vals['25%'],vals['mean'],vals['50%'],vals['mean']+vals['std'],vals['75%'],vals['max']]
    ordenada = sorted(ordenada)
else:
    ordenada = [vals['min'],vals['25%'],vals['mean'],vals['50%'],vals['mean']+vals['std'],vals['75%'],vals['max']]
    ordenada = sorted(ordenada)
for  num,x in enumerate(ordenada):
    try:
        if x <= 0 and ordenada[num+1] > 0:
            ordenada[num+1] = 0
            break
    except:
        pass
    
#ordenada

#traigo los colores que voy a necesitar
positivos = ['#FFF176','#d8ff65','#bbe27e','#9bd467','#70bf5a','#45ab4d','#1a9641'] 
colores = positivos

#Armamos la escala de colores pasando por parametros(colors,index y los extremos (vmin,vmax))
colormaplinear = cm.LinearColormap(colors=colores, index=ordenada,vmin=min(ordenada),vmax=max(ordenada))

scores_series = secciones_map['cantidad_arboles'] #tomamos la columna de electores

def my_color_function(feature):
    #obtenemos el id del la "caracteristica" que queremos para pasar el color a cada poligono
    #esto lo vamos a ver en detalle mas adelante cuando vayamos a agregar la capa de electores
    
    scores = scores_series.get(int(feature['id']), None) 
    
    # aca si el circuito es nulo, devuelvo un color gris que es este '#8c8c8c y sino utiliza nuestra
    # HERMOSA escala de colores que creamos anteriormente
    return '#8c8c8c' if (pd.isnull(scores)) else colormaplinear(scores)

#AGREGO LA CAPA DE ELECTORES PARA LOS CIRCUITOS

# #ejemplo funcion lambda para despues aplicarlo con los colores
# electores_max = lambda electores: max(electores)

# #vamos a generar un vector de electres
# electores_max(electores_circuito_padron['electores'].values)

#Por si llega a pasar algo con el dataframe original, creamos una copia

circ_mapeo = secciones_map.copy()
###########
c_1 = folium.FeatureGroup(name="Comunas",show=True)


c_1.add_child(folium.GeoJson(
                circ_mapeo[['comuna', 'geometry', 'CANTERO OCUPADO', 'CERRADA', 'OCUPADA',
                           'PARCIALMENTE CERRADA', 'SOBREOCUPADA', 'SOBREOCUPADA PARC. CERRADA',
                           'SUBOCUPADA', 'VACÍA', 'altura_arbol_prom',
                           'diametro_altura_pecho_prom', 'cantidad_arboles', 'ancho_acera_prom']], #de nuevo las columnas que vamos a usar
                highlight_function=highlight_function, #esta funcion nos permite que al pasar el mouse se ponga en otro color
                style_function=lambda feature: { #aca usamos una funcion lambda para cada "linea" de dataset
                                                 #automaticamente folium mapea/itera en cada una de las filas para 
                                                 #agregar lo que queremos
                'color':'#00008B', #el color de la linea de contorno
                'fillOpacity': 0.55, #la opacidad de los colores de relleno
                'weight': 1, #el peso de la linea
                'fillColor': my_color_function(feature), #usamos nuestra funcion de colores
                'dashArray': '5, 5' #va a ser una liena punteada
            },
                # seteamos los datos que se quieren mostrar el en tooltip
               tooltip=folium.features.GeoJsonTooltip(
                    fields= ['comuna','cantidad_arboles','ancho_acera_prom','diametro_altura_pecho_prom',
                             'altura_arbol_prom', 'CANTERO OCUPADO', 'CERRADA', 'OCUPADA',
                             'PARCIALMENTE CERRADA', 'SOBREOCUPADA',
                             'SUBOCUPADA', 'VACÍA'], #ponemos que columnas vamos a mostrar
                    aliases = ['Comuna','Arboles(q)','Ancho acera promedio(m)','Diametro al pecho promedio(m)',
                               'Altura promedio(m)','Cantero ocupado(q)','Acera Cerrada(q)','Acera Ocupada(q)',
                               'Acera Parcialmente cerrada(q)','Acera sobreocupada(q)','Acera suboupada(q)','Acera vacia(q)'], #ponemos los nombres con los que se van a ver
                    labels=True,
                    sticky=True,
                    localize=True)

            ))

generador_centro = (i for i in secciones['geometry'].centroid)

generador_etiqueta = (str(i).replace('nan',' Sin Datos.').replace('.0','')  for i in circ_mapeo['comuna'].values)

for c,e in zip(generador_centro,generador_etiqueta):
    try :
        etiqueta = str(format(int(e),",d"))
    except:
        etiqueta = e
    c_1.add_child(folium.Marker(
    location=[c.y,c.x],
    popup=etiqueta,
    icon=folium.DivIcon(html=f"""<div style="font-family: Trebuchet MS; font-size: 13px; color: black">Comuna:{etiqueta}</div>""")
    ))
```

Vamos a ver con que especies trabajar, dado que trabjar con las que mas se repiten puede llegar a no dar, datos tan relevantes

Por lo que en un primer lugar lo que vamos a hacer es sacar un promedio de la altura del arbol por especie, y un promedio del diametro a la altura del pecho tambien por especie

Ademas vamos a contar la cantidad de datos que tenemos para cada especie

Por ultimo, sabiendo que existe una relación entre el alto y el diametro vamos a ordenar los datos y quedarnos con las 5 especies mas altas con cantidad de arboles mayor a 1000 
y con las 5 especies mas bajas con cantidad de arboles mayor a 1000, para poder realizar un analisis robusto

```python
altura_prom = pd.DataFrame(data.groupby(['nombre_cientifico'])['altura_arbol'].mean()).reset_index()
diametro_prom = pd.DataFrame(data.groupby(['nombre_cientifico'])['diametro_altura_pecho'].mean()).reset_index()
count_especies = pd.DataFrame(data.groupby(['nombre_cientifico'])['nombre_cientifico'].count())
count_especies = count_especies.rename(columns={'nombre_cientifico':'cantidad_especie'}).reset_index()
```

```python
totales_especies = altura_prom.merge(diametro_prom,on='nombre_cientifico', how='inner')
totales_especies = totales_especies.merge(count_especies,on='nombre_cientifico', how='inner')
totales_especies_descending = totales_especies.sort_values(by=['altura_arbol'], ascending=False)
```

```python
totales_especies_descending = totales_especies_descending[totales_especies_descending['cantidad_especie'] > 1000].head(5)
totales_especies_descending['altura_arbol'] = totales_especies_descending['altura_arbol'].round(2)
totales_especies_descending['diametro_altura_pecho'] = totales_especies_descending['diametro_altura_pecho'].round(2)
totales_especies_descending['cantidad_especie'] = totales_especies_descending['cantidad_especie'].round(2)
```

```python
totales_especies_descending.to_csv('5masaltas.csv')
```

```python
totales_especies_ascending = totales_especies.sort_values(by=['altura_arbol'], ascending=True)
totales_especies_ascending = totales_especies_ascending[totales_especies_ascending['cantidad_especie'] > 1000]
totales_especies_ascending['altura_arbol'] = totales_especies_ascending['altura_arbol'].round(2)
totales_especies_ascending['diametro_altura_pecho'] = totales_especies_ascending['diametro_altura_pecho'].round(2)
totales_especies_ascending['cantidad_especie'] = totales_especies_ascending['cantidad_especie'].round(2)
```

```python
def add_categorical_legend(folium_map, title, colors, labels):
    if len(colors) != len(labels):
        raise ValueError("colors and labels must have the same length.")

    color_by_label = dict(zip(labels, colors))
    
    legend_categories = ""     
    for label, color in color_by_label.items():
        legend_categories += f"<li><span style='background:{color}'></span>{label}</li>"
        
    legend_html = f"""
    <div id='maplegend' class='maplegend'>
      <div class='legend-title'>{title}</div>
      <div class='legend-scale'>
        <ul class='legend-labels'>
        {legend_categories}
        </ul>
      </div>
    </div>
    """
    script = f"""
        <script type="text/javascript">
        var oneTimeExecution = (function() {{
                    var executed = false;
                    return function() {{
                        if (!executed) {{
                             var checkExist = setInterval(function() {{
                                       if ((document.getElementsByClassName('leaflet-bottom leaflet-left').length) || (!executed)) {{
                                          document.getElementsByClassName('leaflet-bottom leaflet-left')[0].style.display = "flex"
                                          document.getElementsByClassName('leaflet-bottom leaflet-left')[0].style.flexDirection = "column"
                                          document.getElementsByClassName('leaflet-bottom leaflet-left')[0].innerHTML += `{legend_html}`;
                                          clearInterval(checkExist);
                                          executed = true;
                                       }}
                                    }}, 100);
                        }}
                    }};
                }})();
        oneTimeExecution()
        </script>
      """
   

    css = """

    <style type='text/css'>
      .maplegend {
        z-index:9999;
        float:right;
        background-color: rgba(255, 255, 255, 1);
        border-radius: 5px;
        border: 2px solid #bbb;
        padding: 10px;
        font-size:15px;
        font-family: Trebuchet MS;
        positon: relative;
      }
      .maplegend .legend-title {
        text-align: left;
        margin-bottom: 5px;
        font-weight: bold;
        font-size: 90%;
        }
      .maplegend .legend-scale ul {
        margin: 0;
        margin-bottom: 5px;
        padding: 0;
        float: left;
        list-style: none;
        }
      .maplegend .legend-scale ul li {
        font-size: 80%;
        list-style: none;
        margin-left: 0;
        line-height: 18px;
        margin-bottom: 2px;
        }
      .maplegend ul.legend-labels li span {
        display: block;
        float: left;
        height: 16px;
        width: 30px;
        margin-right: 5px;
        margin-left: 0;
        border: 0px solid #ccc;
        }
      .maplegend .legend-source {
        font-size: 80%;
        color: #777;
        clear: both;
        }
      .maplegend a {
        color: #777;
        }
    </style>
    """

    folium_map.get_root().header.add_child(folium.Element(script + css))

    return folium_map
```
